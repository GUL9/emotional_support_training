```python
import cv2 as cv
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
faceCascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
cap = cv.VideoCapture(0)
if not cap.isOpened():
    print("Cannot open camera")
    exit()
```

```python
def mark_face(frame, faces):
    for(x, y, w, h) in faces:
        cv.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
def mark_emotion(frame, emotion):
    cv.putText(frame,
               emotion,
               (50, 50),
               cv.FONT_HERSHEY_SIMPLEX, 3,
               (0, 0, 255),
                2,
                cv.LINE_4)
```

```python
counter = 0
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
        
    # Our operations on the frame come here
    result = DeepFace.analyze(frame, actions = ['emotion'], enforce_detection = False)
    if counter == 0:
        accumulation = result['emotion']
    else:
        for emotion in result['emotion'].keys():
            accumulation[emotion] += result['emotion'][emotion]

    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)
    
    mark_face(frame, faces)
    if counter == 4:
        dominant = max(accumulation, key=accumulation.get)
        mark_emotion(frame, dominant)
        counter = 0
    else:
        counter += 1
    
    # Display the resulting frame
    cv.imshow('frame', frame)
    if cv.waitKey(3) == ord('q'):
        break
# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
        
```
